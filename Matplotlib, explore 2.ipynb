### Background with random number of rectangles with random colours

```python
import matplotlib.pyplot as plt 
import numpy as np 
from matplotlib.patches import Rectangle 
   
data = 200
   
rectangles = [Rectangle(xy = np.random.rand(2) * 10, width = np.random.rand(), height = np.random.rand(), angle = np.random.rand() * 360) 
        for i in range(data)] 
   
fig, ax = plt.subplots(subplot_kw ={'aspect': 'equal'}, figsize=(10,6)) 
  
for i in rectangles: 
    ax.add_artist(i) 
    i.set_clip_box(ax.bbox) 
    i.set_alpha(np.random.rand()) 
    i.set_facecolor(np.random.rand(4)) 

# axes gula
ax.set_xlim(3, 7) 
ax.set_ylim(3, 7) 
   
fig.suptitle('Background', fontweight ="bold") 
  
plt.show() 
```

# Clipping images with patches

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.cbook as cbook


img = plt.imread("blackhole.jpg") 

fig, ax = plt.subplots()
im = ax.imshow(img)
patch = patches.Circle((160, 100), radius=100, transform=ax.transData)
im.set_clip_path(patch)

ax.axis('off')
plt.show()
```

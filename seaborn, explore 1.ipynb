```python
import seaborn as sns
import matplotlib.pyplot as plt
```

###  seaborn offers build-in dataset i.e : iris, titanic etc

```python
iris = sns.load_dataset('iris')
```

```python
iris.head(100)
```

# Distplot

```python
sns.distplot(iris['sepal_length'])
```

```python
sns.distplot(iris['sepal_width'],kde=False,bins=30)
```

```python
sns.distplot(iris['petal_length'])
```

```python
sns.distplot(iris['petal_width'])
```

# jointplot

```python
sns.jointplot(x='sepal_width',y='petal_width',data=iris,kind='scatter')
```

```python
sns.jointplot(x='sepal_width',y='petal_width',data=iris,kind='hex')
```

```python
sns.jointplot(x='sepal_width',y='petal_width',data=iris,kind='reg')
```

# pairplot

```python
sns.pairplot(iris,hue='sepal_width',palette='coolwarm')
```

# kdeplot, rugplot

```python
sns.distplot(iris['sepal_width'],color='b')
sns.color_palette("husl") 
sns.kdeplot(iris['sepal_width'],color='g')
sns.rugplot(iris['sepal_width'],color='r')
```

# stripplot

```python
sns.stripplot(x='species',y='sepal_width',data=iris)
```

# stripplot

```python
sns.swarmplot(x='species',y='sepal_width',data=iris)
```

# barplot

```python
sns.barplot(x='species',y='sepal_width',data=iris)
```

```python
titanic=sns.load_dataset('titanic')
```

```python
sns.barplot(x="sex",y="survived",hue="class", data=titanic)
```

# countplot

```python
sns.countplot(x="deck",palette="Greens_d", data=titanic)
```

# pointplot

```python
sns.pointplot(x="sex",y="survived",
              hue="sex", 
              data=titanic,
              palette={"male":"g", "female":"m"},
              markers=["^","o"],
              linestyles=["-","--"])
```

# boxplot

```python
sns.boxplot(x="alive", y="age", hue="adult_male", data=titanic)
```

```python
 sns.boxplot(data=iris,orient="h") 
```

# violinplot

```python
sns.violinplot(x="age",  y="sex", hue="survived", data=titanic)
```

```python
g = sns.FacetGrid(titanic, col="survived",row="sex")
    
g = g.map(plt.hist,"age")   
```

# factorplot

```python
 sns.factorplot(x="pclass",  y="survived",  hue="sex", data=titanic)
```

# linear model plot

```python
sns.lmplot(x="sepal_width",  y="sepal_length",  hue="species", data=iris)
```

# regplot

```python
sns.regplot(x="sepal_width", y="sepal_length",  data=iris)
```

# distplot

```python
import pandas as pd
import numpy as np

data = pd.DataFrame({'x':np.arange(1,101), 'y':np.random.normal(0,4,100)})

plot = sns.distplot(data.y, kde=True, color="b")
plt.title("A Title") 
plt.ylabel("Survived") 
plt.xlabel("Sex")

ax = plt.axes()
ax.set_facecolor('Aquamarine')
ax.set_alpha(0.15)

plt.show() 
```
